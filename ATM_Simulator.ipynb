```python
class ATM:
    def __init__(self):
        self.balance = 10000  # Initial balance
        self.pin = "65566"  # Default PIN

    def check_pin(self, entered_pin):
        return entered_pin == self.pin

    def display_options(self):
        print("\n***** Welcome To ATM Machine Simulator *****\n")
        print("Options you can Exercise are:")
        print("1) Balance")
        print("2) Withdraw")
        print("3) Deposit")
        print("4) Exit")

    def check_balance(self):
        print("\nYour current balance is: ${}".format(self.balance))

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("\nWithdrawal successful. Remaining balance: ${}".format(self.balance))
        else:
            print("\nInsufficient Balance")

    def deposit(self, amount):
        self.balance += amount
        print("\nDeposit successful. Current balance: ${}".format(self.balance))


def main():
    atm = ATM()

    # Input PIN
    pin = input("Enter Your Pin: ")

    # Validate PIN
    if atm.check_pin(pin):
        while True:
            atm.display_options()

            option = input("\nSelect Your Transaction from the above options: ")

            if option == "1":  # Check Balance
                atm.check_balance()
            elif option == "2":  # Withdraw
                amount = float(input("Enter Amount: "))
                atm.withdraw(amount)
            elif option == "3":  # Deposit
                amount = float(input("Enter Amount: "))
                atm.deposit(amount)
            elif option == "4":  # Exit
                print("\nThank you for using the ATM. Have a nice day!")
                break
            else:
                print("\nInvalid Option! Please select a valid option.")
    else:
        print("\nIncorrect PIN. Access Denied.")


if __name__ == "__main__":
    main()
```
